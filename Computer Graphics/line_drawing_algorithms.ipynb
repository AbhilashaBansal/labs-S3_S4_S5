**DDA Algorithm**

```python
def roundOff(a):
  return int(a + 0.5)

def drawDDA():
    x1 = int(input("Enter intial X coordinate: "))
    y1 = int(input("Enter intial Y coordinate: "))
    x2 = int(input("Enter final X coordinate: "))
    y2 = int(input("Enter final Y coordinate: "))
    
    x,y = x1,y1
    x_coords=[]
    y_coords=[]
    
    length = abs((x2-x1) if (x2-x1) > (y2-y1) else (y2-y1))
    xinc = (x2-x1)/float(length)
    yinc = (y2-y1)/float(length)
    
    print ('x = %s, y = %s' % ( (roundOff(x), roundOff(y)) ))
    x_coords.append(roundOff(x))
    y_coords.append(roundOff(y))

    for i in range(length):
        x += xinc
        y += yinc
        print ('x = %s, y = %s' % ( (roundOff(x), roundOff(y)) ))
        x_coords.append(roundOff(x))
        y_coords.append(roundOff(y))
        
    win = GraphWin("Line using DDA", 500, 500) 
    for i in range(len(x_coords)):
        win.plotPixel(x_coords[i], y_coords[i], "Blue")
        
    win.getMouse () # pause for click in window 
    win.close() 
```

```python
drawDDA()
```

**Bresenham Algorithm**

```python
def drawBresenham():
    x1 = int(input("Enter intial X coordinate: "))
    y1 = int(input("Enter intial Y coordinate: "))
    x2 = int(input("Enter final X coordinate: "))
    y2 = int(input("Enter final Y coordinate: "))
    
    x,y = x1,y1
    x_coords=[]
    y_coords=[]
    dx = abs(x2-x1)
    dy = abs(y2-y1)
    p = 2*dy-dx
    
    print ('x = %s, y = %s' % ( (roundOff(x), roundOff(y)) ))
    x_coords.append(roundOff(x))
    y_coords.append(roundOff(y))
 
    for i in range(dx):
        x = x+1 if x<x2 else x-1

        if p<0:
            p += 2*dy
            
        else:
            p += 2*(dy-dx)
            y = y+1 if y<y2 else y-1
            
        print ('x = %s, y = %s' % ( (roundOff(x), roundOff(y)) ))
        x_coords.append(roundOff(x))
        y_coords.append(roundOff(y))
        
    win = GraphWin("Line using Bresenham", 500, 500) 
    for i in range(len(x_coords)):
        win.plotPixel(x_coords[i], y_coords[i], "Red")
        
    win.getMouse () # pause for click in window 
    win.close() 
```

```python
drawBresenham()
```

**Mid-Point Algorithm**

```python
def drawMidPoint():
    x1 = int(input("Enter intial X coordinate: "))
    y1 = int(input("Enter intial Y coordinate: "))
    x2 = int(input("Enter final X coordinate: "))
    y2 = int(input("Enter final Y coordinate: "))
    
    x,y = x1,y1
    x_coords=[]
    y_coords=[]
    dx = abs(x2-x1)
    dy = abs(y2-y1)
    d = 2*dy-dx
    delta_d = 2*(dy-dx)
    
    print ('x = %s, y = %s' % ( (roundOff(x), roundOff(y)) ))
    x_coords.append(roundOff(x))
    y_coords.append(roundOff(y))
 
    for i in range(dx):
        x = x+1 if x<x2 else x-1
        
        if d<0:
            d += 2*dy
            
        else:
            d += delta_d
            y = y+1 if y<y2 else y-1
            
        print ('x = %s, y = %s' % ( (roundOff(x), roundOff(y)) ))
        x_coords.append(roundOff(x))
        y_coords.append(roundOff(y))
        
    win = GraphWin("Line using Bresenham", 500, 500) 
    for i in range(len(x_coords)):
        win.plotPixel(x_coords[i], y_coords[i], "Blue")
        
    win.getMouse () # pause for click in window 
    win.close() 
```

```python
drawMidPoint()
```
