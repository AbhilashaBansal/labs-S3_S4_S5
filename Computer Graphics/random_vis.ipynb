```python
def drawClouds(sx, sy, win, large=True):
    c1 = Circle(Point(sx,sy), 30)
    c1.setOutline('white')
    c1.setFill('white')
    c1.draw(win)
    c2 = Circle(Point(sx+40,sy-10), 30)
    c2.setOutline('white')
    c2.setFill('white')
    c2.draw(win)
    c3 = Circle(Point(sx+80,sy), 30)
    c3.setOutline('white')
    c3.setFill('white')
    c3.draw(win)
    c4 = Circle(Point(sx+20,sy+20), 30)
    c4.setOutline('white')
    c4.setFill('white')
    c4.draw(win)
    c5 = Circle(Point(sx+60,sy+20), 30)
    c5.setOutline('white')
    c5.setFill('white')
    c5.draw(win)
    
    if large==True:
        c6 = Circle(Point(sx+120,sy), 30)
        c6.setOutline('white')
        c6.setFill('white')
        c6.draw(win)
        c7 = Circle(Point(sx+90,sy+10), 30)
        c7.setOutline('white')
        c7.setFill('white')
        c7.draw(win)
    
```

```python
def fn():
    win = GraphWin('Scene', 1300, 700)
    
    # sky background
    poly = Polygon(Point(0,0), Point(1300,0), Point(1300,700), Point(0, 700))
    poly.setFill(color_rgb(0, 180, 220))
    poly.draw(win)
    
    # clouds
    drawClouds(40, 40, win)
    drawClouds(225, 40, win)
    drawClouds(500, 40, win, False)
    drawClouds(1200, 40, win, False)
    
    # sun
    sun = Circle(Point(1050, 50), 40)
    sun.setOutline('yellow')
    sun.setFill('yellow')
    sun.draw(win)
    
    # tree
    trunk = Rectangle(Point(50, 300), Point(120, 600))
    trunk.setOutline('#795C34')
    trunk.setFill('#795C34')
    trunk.draw(win)
    
    drawTree(win)
    
    # ground
    r1 = Rectangle(Point(0,600), Point(1300, 1000))
    r1.setOutline(color_rgb(225, 173, 1))
    r1.setFill(color_rgb(225, 173, 1))
    r1.draw(win)
    
    time.sleep(3)
    
    # man
    x2 = 1200
    img3 = Image(Point(x2,420), "man1.png")
    img3.draw(win)
    
    # child
    x1 = 350
    img1 = Image(Point(x1,420), "b1.png")
    img1.draw(win)
    
    for i in range(8):
        time.sleep(0.5)
        img1.undraw()
        img3.undraw()
        if i%2==0:
            img1 = Image(Point(x1+ ((i+1)*15), 420), "b2.png")
            img1.draw(win)
            img3 = Image(Point(x2- ((i+1)*15), 420), "man2.png")
            img3.draw(win)
        else:
            img1 = Image(Point(x1+((i+1)*15), 420), "b1.png")
            img1.draw(win)
            img3 = Image(Point(x2- ((i+1)*15), 420), "man1.png")
            img3.draw(win)
        
        i+=1
        
    
    # dialogue boxes
    
    dr1 = Rectangle(Point(450, 125), Point(750, 225))
    dr1.setFill('#fffbc8')
    dr1.draw(win)
    
    dr2 = Rectangle(Point(900, 125), Point(1200, 225))
    dr2.setFill('#fffbc8')
    dr2.draw(win)
    
    # dialogues
    # 1
    d1 = Text(Point(600, 175), "Good afternoon sir! How are you?")
    d1.setSize(18)
    d1.draw(win)
    time.sleep(2)
    
    # 2
    d2 = Text(Point(1050, 175), "I am fine, how are you?")
    d2.setSize(18)
    d2.draw(win)
    time.sleep(2)
    
    d1.undraw()
    d2.undraw()
    
    # 3
    d11 = Text(Point(600, 150), "I am fine too.")
    d12 = Text(Point(600, 175), "Sir, I have heard about the CORONA Virus.")
    d13 = Text(Point(600, 200), "I am worried!")
    d11.setSize(16)
    d12.setSize(16)
    d13.setSize(16)
    d11.draw(win)
    d12.draw(win)
    d13.draw(win)
    time.sleep(3)
    
    # 4
    d21 = Text(Point(1050, 140), "There is no need to worry.")
    d22 = Text(Point(1050, 160), "Just follow 3 simple steps.")
    d23 = Text(Point(1050, 190), "Wear a mask, maintain 6 feet distance,")
    d24 = Text(Point(1050, 210), "and keep your hands & face clean.")
    d21.setSize(16)
    d22.setSize(16)
    d23.setSize(16)
    d24.setSize(16)
    d23.setStyle('bold')
    d24.setStyle('bold')
    d21.draw(win)
    d22.draw(win)
    time.sleep(3)
    d23.draw(win)
    d24.draw(win)
    time.sleep(5)
    
    d11.undraw()
    d12.undraw()
    d13.undraw()
    d21.undraw()
    d22.undraw()
    d23.undraw()
    d24.undraw()
    
    # 5
    d11 = Text(Point(600, 150), "Okay, thank you sir!")
    d12 = Text(Point(600, 175), "I will take care of these things")
    d13 = Text(Point(600, 200), "Bye sir.")
    d11.setSize(16)
    d12.setSize(16)
    d13.setSize(16)
    d11.draw(win)
    d12.draw(win)
    d13.draw(win)
    time.sleep(2)
    
    d2 = Text(Point(1050, 175), "Great, bye bye!")
    d2.setSize(18)
    d2.draw(win)
    time.sleep(2)
    
    d11.undraw()
    d12.undraw()
    d13.undraw()
    d2.undraw()
    
    dr1.undraw()
    dr2.undraw()
    
    
    # man, child walk back
    for i in range(8, 45):
        time.sleep(0.5)
        img1.undraw()
        img3.undraw()
        if i%2==0:
            img1 = Image(Point(x1+ ((i+1)*20), 420), "b2.png")
            img1.draw(win)
            img3 = Image(Point(x2- ((i+1)*20), 420), "man2.png")
            img3.draw(win)
        else:
            img1 = Image(Point(x1+((i+1)*20), 420), "b1.png")
            img1.draw(win)
            img3 = Image(Point(x2- ((i+1)*20), 420), "man1.png")
            img3.draw(win)
        
        i+=1
        
    
    win.getMouse()
    win.close()
```
